```python
#Crear una serie a partir de una lista
import pandas as pd
lista = ["Bogotá","Cali","Medellin","Cartagena","Bucaramanga"]
serie = pd.Series(lista, name = "Ciudad") # Crear Serie
print (serie)

df_ciudad = pd.DataFrame(data = lista, columns= ["Ciudad"])#Creamos una sola columna - Df = DataFrame
df_ciudad
```

```python
df_ciudad.info() #Descripción del DF
```
