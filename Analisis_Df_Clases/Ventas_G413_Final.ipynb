```python
#Cambiar nombre a una Columna

df_ventas = df_ventas.rename(columns={"ID_Pedido":"Pedido","Zona":"Continente"})
df_ventas
```

```python
#Cambiar nombre de indice

df_ventas= df_ventas.rename(index={"C1069":"C1000"})
df_ventas
```

```python
#Tratamiento de datos faltantes o nulos isna() isnull()

df_ventas.isna().sum() #datos faltantes
```

```python
#Imputación de datos fatantes con .fillna()busca los valores faltantes y los modifica por los que están dentro de los parentesis

df_ventas.Unidades = df_ventas.Unidades.fillna(df_ventas.ImporteVentaTotal/df_ventas.Precio_Unitario)
df_ventas
```

```python
df_ventas.isna().sum()
df_ventas.info()
```

```python
#Cambiar tipo de dato de una columna

df_ventas.Unidades = df_ventas.Unidades.astype('int')
df_ventas.info()
```

```python
#Cambiar tipo de dato a Coste Unitario

#df_ventas.Coste_Unitario = df_ventas.Coste_Unitario.astype('float') - sale error porque encontre un string

#Conversion Forzada
df_ventas.Coste_Unitario = pd.to_numeric(df_ventas.Coste_Unitario, errors='coerce')
df_ventas.info()
df_ventas.isnull().sum()
df_ventas
```

```python
#Convertir la columna ImporteCosteTotal a float
#df_ventas.ImporteCosteTotal = df_ventas.ImporteCosteTotal.astype("float") - 
# No convierte a float por que hay string

#Forzamos la conversión
df_ventas.ImporteCosteTotal = pd.to_numeric(df_ventas.ImporteCosteTotal, errors="coerce")
df_ventas.info()
```

```python
#Imputacion de ImporteCosteTotal

df_ventas.ImporteCosteTotal = df_ventas.ImporteCosteTotal.fillna(df_ventas.Unidades*df_ventas.Coste_Unitario)
df_ventas.info()
```

```python
#Imputar Coste Unitario

df_ventas.Coste_Unitario = df_ventas.Coste_Unitario.fillna(df_ventas.ImporteCosteTotal/df_ventas.Unidades)
df_ventas
```

```python
#Imputar Fecha de Envio: Diferencia entre fechaEnvio y FechaPedido, dias
#Calculamos el promedio de los dia
#FechaEnvio=FechaPedio+promedio

df_ventas['Días'] = df_ventas.Fecha_Envío - df_ventas.Fecha_Pedido
promedioDias=df_ventas.Días.mean() #Calcula promedio - media
print(promedioDias)
```

```python
#Imputar Fecha_Envio

df_ventas.Fecha_Envío = df_ventas.Fecha_Envío.fillna(df_ventas.Fecha_Pedido + promedioDias)
df_ventas.isna().sum()
```

```python
#Imputar Fecha_Pedido
df_ventas.Fecha_Pedido = df_ventas.Fecha_Pedido.fillna(df_ventas.Fecha_Envío - promedioDias)
df_ventas.isna().sum()
```

```python
#Consultar el cliente donde falta el id del pedido
df_ventas[df_ventas.Pedido.isna()]
```

```python
#Eliminar el registro asociado a ese cliente
df_ventas = df_ventas.drop(index='C3931')
df_ventas.info()
```

```python
#Encontrar datos duplicados

df_ventas.duplicated().sum() #Consultar datos duplicados en el df - filas iguales
df_ventas[df_ventas.duplicated()] #Mostrar las filas duplicadas

#Duplicados en una Columna
df_ventas[df_ventas.duplicated(subset="Prioridad")] #Mostrar duplicados en columnas
```

```python
#Eliminar duplicados usando .drop_duplicates

df_ventas.drop_duplicates(keep="first") #Elimina filas duplicadas - dejando la primera fila

#keep = "first" Deja la primer fila y el resto las elimina
#Keep = "last" Deja la ultima fila y el resto las elimina
#keep = False Elimina todas las filas duplicadas.
```

```python
#Eliminar columna Días
df_ventas = df_ventas.drop("Días", axis=1) #axis=0 elimina columna  // axis = 1 elimina fila. 
df_ventas.info()
```
